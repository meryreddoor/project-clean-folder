```python
#Importing all the relevant libraries.
import pandas as pd
import numpy as np
import re
import matplotlib
import matplotlib.pyplot as plt
```

```python
#RAW FOLDER - Shark Attack
shark_csv_folder = pd.read_csv("OUTPUT/GSAF5.csv",encoding='latin1')
shark_csv_folder
```

```python
#Removing two last columns. They do not have relevant information for our purpose.
shark_csv_folder=shark_csv_folder.drop(['Unnamed: 22', 'Unnamed: 23'], axis=1)
```

```python
#Checking what is the type of the columns in order to know how can we work on them.
shark_csv_folder.dtypes
```

```python
#Removing whitespaces from header.
shark_csv_folder=shark_csv_folder.rename(columns=lambda x: x.strip())
```

```python
#Checking if the white spaces have been removed from header. 
shark_csv_folder.columns
```

```python
#Removing white spaces from Country Column.
shark_csv_folder['Country']=shark_csv_folder['Country'].str.strip()
```

```python
#Removing ? from Country Column.
shark_csv_folder['Country']=shark_csv_folder['Country'].str.replace("?",'')
```

```python
#Removing the empty cells. 
shark_csv_folder=shark_csv_folder.fillna(0)
```

```python
#Checking what are the remaining null values.
shark_csv_folder=shark_csv_folder.isnull().sum()[shark_csv_folder.isnull().sum()>0]
```

```python
#Removing the remaining null values.
shark={'Country':'OUT OF SCOPE'}
shark_csv_folder.fillna(value=shark,inplace=True)
```

```python
#Removing white spaces from Year Column.
shark_csv_folder["Year"] = shark_csv_folder["Year"].astype(str)
shark_csv_folder['Year']= shark_csv_folder['Year'].str.strip()
```

```python
#Cleaning Year Column.
for e in shark_csv_folder['Year']:
    if e <1300:
        shark_csv_folder['Year']=shark_csv_folder['Year'].replace(e, 'OUT OF SCOPE')
```
