```python
#(“High Human Development”), with more than 0,80
#(“Medium Human Development”), from 0,70 to 0,80
#(“Medium Human Development”), from 0,50 to 0,70
#(“Low Human Development”), with less than 0,55
```

```python
Human_Development = []

for x in hdi['Year 2018']:
    
    if x >= 0.8:
        Human_Development.append('High Human Development')
    elif 0.8 > x >= 0.5:
        Human_Development.append('Medium Human Development')
    else:
        Human_Development.append('Low Human Development')

hdi['Human_Development'] = Human_Development
```
