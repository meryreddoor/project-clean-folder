```python
import pandas as pd
shark_csv_folder = pd.read_csv("OUTPUT/v2_GSAF5.csv",encoding='latin1')
shark_csv_folder.head()
```

```python
for i in range(len(shark_csv_folder)):
    dicc1={'NEW BRITAIN':'PAPUA NEW GUINEA',
           'ENGLAND':'UNITED KINGDOM', 
           'SCOTLAND':'UNITED KINGDOM',
           'WESTERN SAMOA':'SAMOA',
           'ST. MAARTIN':'SAINT MARTIN', 
           'ST. MARTIN':'SAINT MARTIN',
           'ST. HELENA':'SAINT HELENA',
           'ST HELENA':'SAINT HELENA'
          }
    
    for x,y in dicc1.items(): 
            shark_csv_folder['Country'].values[i] = shark_csv_folder['Country'].values[i].replace(x, y)

shark_csv_folder['Country']
```
